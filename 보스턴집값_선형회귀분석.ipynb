```python
import matplotlib as mpl
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np
import scipy as sp
import pandas as pd
import statsmodels.api as sm
import sklearn as sk
```

```python
from sklearn.datasets import load_boston

boston = load_boston()

dfX0 = pd.DataFrame(boston.data, columns=boston.feature_names)
dfX = sm.add_constant(dfX0)
dfy = pd.DataFrame(boston.target, columns=["MEDV"])
```

## 1. scikit-learn 패키지를 사용

```python
from sklearn.linear_model import LinearRegression

# 사이킷런 패키지 방법에서는 사용자가 상수항 결합을 해줄 필요가 없다.
model = LinearRegression().fit(dfX0, dfy)
print(model.intercept_, model.coef_)
```

## 2. statsmodels 패키지를 사용


2.1. OLS 클래스로만 하는 방법

```python
# 상수항을 결합해서 인수로 넣어야 한다.
model2 = sm.OLS(dfy, dfX)
```

```python
result2 = model2.fit()
print(result2.summary())
```

2.2. OLS 클래스의 from_formula 메소드로 하는 방법

```python
df = pd.concat([dfX, dfy], axis=1)
df
```

```python
model3 = sm.OLS.from_formula("MEDV ~ " + "+".join(boston.feature_names), data=df)
result3 = model3.fit()
```

```python
print(result3.summary())
```

3가지 방법으로 도출한 계수(coefficient)와 상수항이 모두 동일함을 알 수 있다.
