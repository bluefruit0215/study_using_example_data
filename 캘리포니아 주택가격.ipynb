```python
import matplotlib as mpl
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np
import scipy as sp
import pandas as pd
import statsmodels.api as sm
import sklearn as sk
```

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df["Price"] = housing.target
df.tail()
```

```python
model = sm.OLS.from_formula("Price ~ " + "+".join(housing.feature_names), data=df)
result = model.fit()
```

```python
print(result.summary())
```

조건수가 매우 크니 스케일링을 한 후에 다시 선형회귀분석을 수행한다

```python
feature_names = ["scale({})".format(name) for name in list(housing.feature_names)]
model = sm.OLS.from_formula("Price ~ " + "+".join(feature_names), data=df)
result = model.fit()
```

```python
print(result.summary())
```

- 조건수가 6.67로 낮아진 것을 알 수 있다.
- 독립변수 Pouplation의 유의확률이 약 40.2%로 Population의 계수가 0이다라는 귀무가설을 채택한다. 즉, Population은 집값을 구별하는데 의미가 없다. 이를 부분회귀 플롯으로 확인해본다.

```python
fig = plt.figure(figsize=(10, 15))
sm.graphics.plot_partregress_grid(result, fig=fig)
plt.show()
```

부분회귀 플롯으로도 Population 독립변수가 종속변수와 상관관계가 없음을 알 수 있다.  
Population 독립변수를 제외하고 선형회귀분석을 다시 수행해본다.

```python
feature_names.remove("scale(Population)")
model = sm.OLS.from_formula("Price ~ " + "+".join(feature_names), data=df)
result = model.fit()
```

```python
print(result.summary())
```
